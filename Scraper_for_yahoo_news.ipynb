```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd

                    ##______________ Start Selenium setup______________________##
    

chromedriver_location = 'F:\Data Science Projects\chromedriver.exe'  #Use your chromedriver.exe filepath
browser = webdriver.Chrome(executable_path=chromedriver_location) 
                                                          
#browser = webdriver.Chrome(ChromeDriverManager().install())      #uncheck this line if you don't have chromedrive

url = "https://finance.yahoo.com/quote/INFY/news?p=INFY"
browser.get(url)
time.sleep(1)

element = browser.find_element_by_tag_name("body")

no_of_pagedowns = 100

while no_of_pagedowns:
    element.send_keys(Keys.PAGE_DOWN)
    time.sleep(0.2)
    no_of_pagedowns-=1


page_source = browser.page_source      

                                ###_________End of selenium part_____________ ##
                                ###_________Beautifulsoup Beginning___________##

soup = BeautifulSoup(page_source)

post_title = soup.findAll("h3", {"class": "Mb(5px)"})

Titles = []
for item in post_title:
    Titles.append(item.text)

date_section = soup.findAll("div", {"class": "C(#959595) Fz(11px) D(ib) Mb(6px)"})

dates = []
for item in date_section :
    dates.append(item.text.split("•")[1])
    
               ###_____________Creating DataFrame using the scrapped Data__________###
        

df = pd.DataFrame()
df['post_title'] = Titles
df['time_of_posting'] = dates
```
